## HIV-HNRC

```python
#read-in HNRC-HIV data
hiv_feature = pd.read_csv('../data/HNRC_HIV_feature_table_filtered.tsv', sep='\t', index_col=0)
hiv_metabolomics = pd.read_csv('../data/HNRC_HIV_metabolomics_ft.csv', index_col=0)
hiv_metadata = pd.read_csv('../data/HNRC_HIV_metadata.tsv', sep='\t', index_col=0)
```

```python
#match index order of metadata
hiv_feature = hiv_feature.loc[hiv_metadata.index]
hiv_metabolomics = hiv_metabolomics.loc[hiv_metadata.index]
```

```python
print(hiv_feature.shape)
display(hiv_feature.head())
print(hiv_metabolomics.shape)
display(hiv_metabolomics.head())
print(hiv_metadata.shape)
display(hiv_metadata.head())
```

## U19-Wisconsin

```python
#read-in U19-Wisconsin data
u19_feature = pd.read_csv('../data/U19_Wisconsin_feature_table_filtered.tsv', sep='\t', index_col=0)
u19_metabolomics = pd.read_csv('../data/U19_Wisconsin_metabolomics_ft.csv', index_col=0)
u19_metadata = pd.read_csv('../data/U19_Wisconsin_matched_metadata.tsv', sep='\t', index_col=0)
```

```python
#match index order of metadata
u19_feature = u19_feature.loc[u19_metadata.index]
#add 13663 to id names
u19_metabolomics.index = ['13663.{}'.format(i) for i in u19_metabolomics.index]
u19_metabolomics = u19_metabolomics.loc[u19_metadata.index]
```

```python
print(u19_feature.shape)
display(u19_feature.head())
print(u19_metabolomics.shape)
display(u19_metabolomics.head())
print(u19_metadata.shape)
display(u19_metadata.head())
```

```python
#save formatted data
hiv_feature.to_csv('../data/HNRC_HIV_feature_table_formatted.tsv', sep='\t')
hiv_metabolomics.to_csv('../data/HNRC_HIV_metabolomics_ft_formatted.csv')
u19_feature.to_csv('../data/U19_Wisconsin_feature_table_formatted.tsv', sep='\t')
u19_metabolomics.to_csv('../data/U19_Wisconsin_metabolomics_ft_formatted.csv')
```
